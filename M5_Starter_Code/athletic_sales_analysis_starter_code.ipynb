```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
sales_2020_df
```

```python
# Display the 2021 sales DataFrame
sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([sales_2020_df, sales_2021_df], axis='rows', join='inner').reset_index(drop=True)
combined_df
```

```python
# Check if any values are null.
combined_df.isnull().sum()
```

```python
# Check the data type of each column
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'], format='%m/%d/%y')
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df['invoice_date'].dtype
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
tot = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
tot = tot.rename(columns={'units_sold': 'Total_Products_Sold'}).sort_values('Total_Products_Sold', ascending=False)



# Show the top 5 results.
tot.head()                           
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
tot = pd.pivot_table(combined_df,
                     index=['region', 'state', 'city'],
                     values='units_sold',
                     aggfunc='sum').sort_values('units_sold', ascending=False)

# Rename the "units_sold" column to "Total_Products_Sold"
tot = tot.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
tot.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
tot_sales = combined_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
tot_sales = tot_sales.rename(columns={'total_sales': 'Total_Sales'}).sort_values('Total_Sales', ascending=False)

# Show the top 5 results.
tot_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
tot_sales = pd.pivot_table(combined_df,
                     index=['region', 'state', 'city'],
                     values='total_sales',
                     aggfunc='sum').sort_values('total_sales', ascending=False)


# Optional: Rename the "total_sales" column to "Total Sales"
tot_sales = tot_sales.rename(columns={'total_sales': 'Total_Sales'})

# Show the top 5 results.
tot_sales.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
ret_sales = combined_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()
ret_sales = ret_sales.rename(columns={'total_sales': 'Total_Sales'}).sort_values('Total_Sales', ascending=False)
# Show the top 5 results.
ret_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
ret_sales = pd.pivot_table(combined_df,
                     index=['retailer', 'region', 'state', 'city'],
                     values='total_sales',
                     aggfunc='sum').sort_values('total_sales', ascending=False)

# Optional: Rename the "total_sales" column to "Total Sales"
ret_sales = ret_sales.rename(columns={'total_sales': 'Total_Sales'})

# Show the top 5 results.
ret_sales.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_df = combined_df[combined_df['product'].str.contains("Women's", case=False)]
womens_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
wom_tot = womens_df.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()
wom_tot = wom_tot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}).sort_values('Womens_Footwear_Units_Sold', ascending=False)
# Show the top 5 results.
wom_tot.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
wom_tot = pd.pivot_table(womens_df,
                     index=['retailer', 'region', 'state', 'city'],
                     values='units_sold',
                     aggfunc='sum').sort_values('units_sold', ascending=False)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold" 
wom_tot = wom_tot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
# Show the top 5 results.
wom_tot.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_df.loc[:, 'invoice_date'] = pd.to_datetime(womens_df['invoice_date'], format='%m/%d/%y')
wom_in_data = pd.pivot_table(womens_df,
                     index='invoice_date',
                     values='total_sales',
                     aggfunc='sum')
# Optional: Rename the "total_sales" column to "Total Sales"
wom_in_data = wom_in_data.rename(columns={'total_sales': 'Total_Sales'})


# Show the table.
wom_in_data
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = wom_in_data.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales = daily_sales.sort_values('Total_Sales', ascending=False)
daily_sales.head(20)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = wom_in_data.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_sales = weekly_sales.sort_values('Total_Sales', ascending=False)
weekly_sales.head(20)
```
